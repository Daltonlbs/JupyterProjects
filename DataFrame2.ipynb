```python
#você pode realizar a mesma ação com o dropna()

fbok3 = fbok.dropna()
fbok3.shape
```

```python
#fbok4 = fbok.dropna(axis=0); axis 0 é o default da função
#Caso queira eliminar as colunas ao invés das linas pode-se utilizar axis=1

fbok4 = fbok.dropna(axis=1)
fbok4.shape
```

```python
#Geralmente as informações nulas são substituidas

#Preenchendo com zeros
fbok.fillna(0)

#Preenchendo com o valor anterior
fbok.fillna(method='pad')

#Ou o valor posterior
fbok.fillna(method='bfill')

#Últimos dois métodos são raramente utilizados
```

```python
fbok["share"].fillna(-99, inplace = True)
```

```python
#Um método mais sofisticado é estimando uma regressão dos adados existentes para estimar os faltantes

fbok5 = fbok.interpolate(methos = 'linear', limit_direction='forward')

#Este é um dos métodos masi utilizados para o tratamento de missing values
fbok5.isnull().sum()
```
