```python
XX=dol

BR1Y = pd.DataFrame()
BR1Y["BR1Y"] = inv.get_bond_historical_data('Brazil 1Y', from_date=inicio, to_date =hoje)['Close']
BR1Y.index.name = "Date"
BR1Y.reset_index(inplace = True)
BR1Y["Date"] = pd.to_datetime(BR1Y["Date"])
XX=pd.merge(XX, BR1Y,on = "Date", how='outer')

BR3Y = pd.DataFrame()
BR3Y["BR3Y"] = inv.get_bond_historical_data('Brazil 3Y', from_date=inicio, to_date =hoje)['Close']
BR3Y.index.name = "Date"
BR3Y.reset_index(inplace = True)
BR3Y["Date"] = pd.to_datetime(BR3Y["Date"])
XX=pd.merge(XX, BR3Y,on = "Date", how='outer')

BR10Y = pd.DataFrame()
BR10Y["BR10Y"] = inv.get_bond_historical_data('Brazil 10Y', from_date=inicio, to_date =hoje)['Close']
BR10Y.index.name = "Date"
BR10Y.reset_index(inplace = True)
BR10Y["Date"] = pd.to_datetime(BR10Y["Date"])
XX=pd.merge(XX, BR10Y,on = "Date", how='outer')

US3M = pd.DataFrame()
US3M["US3M"] = inv.get_bond_historical_data('U.S. 3M', from_date=inicio, to_date =hoje)['Close']
US3M.index.name = "Date"
US3M.reset_index(inplace = True)
US3M["Date"] = pd.to_datetime(US3M["Date"])
XX=pd.merge(XX, US3M,on = "Date", how='outer')

US5Y = pd.DataFrame()
US5Y["US5Y"] = inv.get_bond_historical_data('U.S. 5Y', from_date=inicio, to_date =hoje)['Close']
US5Y.index.name = "Date"
US5Y.reset_index(inplace = True)
US5Y["Date"] = pd.to_datetime(US5Y["Date"])
XX=pd.merge(XX, US5Y,on = "Date", how='outer')
                                            
US10Y = pd.DataFrame()
US10Y["US10Y"] = inv.get_bond_historical_data('U.S. 10Y', from_date=inicio, to_date =hoje)['Close']
US10Y.index.name = "Date"
US10Y.reset_index(inplace = True)
US10Y["Date"] = pd.to_datetime(US10Y["Date"])
XX=pd.merge(XX, US10Y,on = "Date", how='outer')

BRENT = pd.DataFrame()
BRENT["BRENT"] = inv.get_commodity_historical_data( 'Brent Oil', from_date=inicio, to_date =hoje)['Close']
BRENT.index.name = "Date"
BRENT.reset_index(inplace = True)
BRENT["Date"] = pd.to_datetime(BRENT["Date"])
XX=pd.merge(XX, BRENT,on = "Date", how='outer')

OURO = pd.DataFrame()
OURO["OURO"] = inv.get_commodity_historical_data( 'Gold', from_date=inicio, to_date =hoje)['Close']
OURO.index.name = "Date"
OURO.reset_index(inplace = True)
OURO["Date"] = pd.to_datetime(OURO["Date"])
XX=pd.merge(XX, OURO,on = "Date", how='outer')

dxy = pd.DataFrame()
dxy["DXY"] = inv.get_index_historical_data('US Dollar Index', from_date=inicio, country = "United States", to_date =hoje)['Close']
dxy.index.name = "Date"
dxy.reset_index(inplace = True)
dxy["Date"] = pd.to_datetime(dxy["Date"])
XX=pd.merge(XX, dxy,on = "Date", how='outer')

mxn = pd.DataFrame()
mxn["MXN"] = inv.get_currency_cross_historical_data('USD/MXN', from_date=inicio, to_date ='01/09/2021')['Close']
mxn.index.name = "Date"
mxn.reset_index(inplace = True)
mxn["Date"] = pd.to_datetime(mxn["Date"])
XX=pd.merge(XX, mxn,on = "Date", how='outer')

TRY = pd.DataFrame()
TRY["TRY"] = inv.get_currency_cross_historical_data('USD/TRY', from_date=inicio, to_date ='01/09/2021')['Close']
TRY.index.name = "Date"
TRY.reset_index(inplace = True)
TRY["Date"] = pd.to_datetime(TRY["Date"])
XX=pd.merge(XX, TRY,on = "Date", how='outer')

GBAD = pd.DataFrame()
GBAD["GBAD"] = inv.get_currency_cross_historical_data('GBP/AUD', from_date=inicio, to_date ='01/09/2021')['Close']
GBAD.index.name = "Date"
GBAD.reset_index(inplace = True)
GBAD["Date"] = pd.to_datetime(GBAD["Date"])
XX=pd.merge(XX, GBAD,on = "Date", how='outer')
```

```python
XX=XX.sort_values(by=['Date'])
```

```python
XX["Date"] = pd.to_datetime(XX["Date"]).dt.normalize()
XX = XX.set_index("Date")
XX = XX.replace(0, np.NaN)
XX = XX.fillna(method = "ffill")
XX
```
