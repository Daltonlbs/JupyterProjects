```python
def RemoveStopWords(inst):
    stopwords = set(nltk.corpus.stopwords('portuguese'))
    palavras = [i for i in inst.split() if not i in stopwords]
    return (" ".join(palavras))
```

```python
def Stemming(inst):
    stemmer = nltk.stem.RSLPStemmer()
    palavras = []
    for w in inst.split():
        palavras.append(stemmer(w))
    return(" ".join(palavras))
```

```python
def Preprocessing(inst):
    inst = re.sub(r"http\S+", "", inst).lower().replace('.','').replace(';','').replace('-','').replace(':','').replace(')','')
    stopwords = set(nltk.corpus.stopwords.words('portuguese'))
    palavras = [i for i in inst.split() if not i in stopwords]
    return (" ".join(palavras))
```
