```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline

#Se houver algum erro com a biblioteca Pandas como o abaixo, instale a vers√£o 0.25
# UnboundLocalError: local variable 'arith_flex' referenced before assignment
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

```python
dados = np.linspace(0,5,50)

dados

cos = np.cos(dados)
arctan = np.arctan(dados)
```

```python
dados
```

```python
fig = plt.figure()  #permite dois gr√°ficos na mesma figura
plt.plot(dados,cos,"-")
plt.plot(dados,arctan, "--"); # ; elimina o linha de especifica√ß√£o do objeto
```

```python
plt.figure()
plt.subplot(2,1,1)    #linhas, colunas e painel
plt.plot(arctan)
plt.subplot(2,1,2)
plt.plot(cos);
```

```python
#Modificando a exibi√ß√£o do subplot

plt.figure()
plt.subplot(1,2,1)    #linhas, colunas e painel
plt.plot(arctan)
plt.subplot(1,2,2)
plt.plot(cos);
```

```python
#ajustando os limites do eixos do gr√°fico e incluindo a legenda

plt.plot(dados, label = "Meu vetor")
plt.plot(dados -2, label = "Meu vetor deslocado")
plt.xlim(0,50)
plt.ylim(-5,8)
plt.legend()
plt.title("Meu gr√°fico")
```

```python
#Como o Python √© uma linguagem orientada √† objetos, voc√™ pode trabalhar desta forma tamb√©m para atribuir oplot
fig, ax = plt.subplots(2)

ax[0].plot(arctan, color = "red")
ax[1].plot(cos, color = "g");


ax[0].set(xlim=(0,50), ylim=(-2,2)
            , xlabel = "x", ylabel = "Valores"
            , title = "Arco Tangente")
ax[1].set(xlim=(0,50), ylim=(-2,2)
            , xlabel = "x", ylabel = "Outros valores"
            , title = "Cosseno")

# E este aqui √© para afastar os gr√°ficos e melhorar a visualiza√ß√£o
fig.tight_layout()

# e para mostrar a legenda
ax[0].legend(['Meu vetor'])
ax[1].legend(['Meu outro vetor']);
```

```python
#SCATTER PLOTS - GR√ÅFICO DE DISPERS√ÉO
#s√£o necess√°rios sempre 2 vetores, x e y, de mesma dimens√£o

randx = np.random.random(75)
randy = np.random.random(75)
```

```python
plt.plot(randx,randy,"*", color = "black", label = "aleat√≥rio") #substitui "." por "*"
plt.legend()
plt.title("Dispers√£o aleat√≥roa")
```

```python
#HISTOGRAMAS

rand = np.random.randn(1000)

plt.hist( rand, bins= 30
        , alpha = 0.5
        , histtype = "stepfilled"
        , color="darkgreen"
        , edgecolor = "red"#"none"
        , label = "Mais aleat√≥rio")

plt.legend()
plt.title("Distribui√ß√£o aleat√≥ria");
```

```python
#diversos histogramas em apenas 1 gr√°fico

dist1 = np.random.normal(4, 0.75,500) # m√©dia, desvio padr√£o, quantidade de elementos
dist2 = np.random.normal(-3, 1,500)
dist3 = np.random.normal(0.5, 2,500)
dist4 = np.random.normal(0.5,0.5,500)

#Vamos criar um dicion√°rio com a configura√ß√£o, para evitar repetir c√≥digo

config = dict(histtype = 'stepfilled', alpha = 0.35, density = True, bins = 50)

plt.hist(dist1, **config, label = "Dist 1")
plt.hist(dist2, **config, label = "Dist 2")
plt.hist(dist3, **config, label = "Dist 3")
plt.hist(dist4, **config, label = "Dist 4")

#formata√ß√£o da legenda - v√°lida para qualquer tipo de gr√°fico

plt.legend (loc = "upper left" #localiza√ß√£o
           , frameon = True    #moldura
           , ncol = 2         #2 colunas
           , fancybox = True   #modifica√ß√£o no quadro
           , framealpha = 0.95  #transpar√™ncia
           , shadow = True   #sombra
           , borderpad = 1)   #espessura da borda

plt.title("Diversas distribui√ß√µes");
```

```python
#O grafico do Ibovespa mudando estilos com style.context

with plt.style.context("seaborn-deep"):
    plt.figure(figsize = (18,10))
    plt.plot(bvsp["Data"], bvsp["Close"], label = "Ibovespa" , linewidth = 1)
    plt.xlabel("2001 √† 2019")
    plt.ylabel("Pontua√ß√£o")
    plt.legend(loc = "upper left"
              , frameon = True
              , fancybox = True
              , framealpha = 1
              , borderpad = 1)
    plt.title("Gr√°fico do Ibov")
```
